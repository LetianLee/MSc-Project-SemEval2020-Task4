## 1. General Preparation

Import some common libraries.

```python
from tqdm import tqdm
import time
```

Use GPU Facilities.

```python
import torch
device = torch.device("cuda:2" if torch.cuda.is_available() else "cpu")
```

## 2. Data processing

### 2.1 Read data from csv

Build a common function to get texts and labels from csv file.

```python
import pandas as pd
def get_info_from_csv(texts_path, labels_path):
    texts = pd.read_csv(texts_path).fillna(str())
    labels = pd.read_csv(labels_path, header=None, names=['id', 'ans'])['ans']
    return texts, labels
```

Read texts and labels from csv file.

```python
train_texts, train_labels = get_info_from_csv(
    'ALL data/Training Data/subtaskB_data_all.csv',
    'ALL data/Training Data/subtaskB_answers_all.csv'
)

val_texts, val_labels = get_info_from_csv(
    'ALL data/Dev Data/subtaskB_dev_data.csv',
    'ALL data/Dev Data/subtaskB_gold_answers.csv'
)

test_texts, test_labels = get_info_from_csv(
    'ALL data/Test Data/subtaskB_test_data.csv',
    'ALL data/Test Data/subtaskB_gold_answers.csv'
)
```

### 2.2 Tokenization

```python
train_texts.head()
```

```python
from transformers import DistilBertTokenizerFast

# Get tokenizer
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')

# Tokenization
# train_encodings = tokenizer([[list(train_texts["FalseSent"]), list(train_texts["OptionA"])], [list(train_texts["FalseSent"]), list(train_texts["OptionB"])], [list(train_texts["FalseSent"]), list(train_texts["OptionC"])]], truncation=True, padding=True)
# val_encodings = tokenizer(list(val_texts["sent0"]), list(val_texts["sent1"]), truncation=True, padding=True)
# test_encodings = tokenizer(list(test_texts["FalseSent"]), list(test_texts["OptionA"]), truncation=True, padding=True)
# test_encodings
```

```python

input_ids_list = list()
attention_mask_list = list()
for text in train_texts.values:
#     print((text))
    encoding = tokenizer([[text[1], text[2]], [text[1], text[3]], [text[1], text[4]]], truncation="longest_first", padding="max_length", max_length =128)
#     print(encoding)
    input_ids_list.append(encoding.input_ids)
    attention_mask_list.append(encoding.attention_mask)
    
train_encodings = {"input_ids":input_ids_list,"attention_mask":attention_mask_list}

```

```python
# train_labels.head()
# print(train_labels)
train_labels = [ord(train_label) - 65 for train_label in train_labels]
```

### 2.3 Turn data into a Dataset object

Define a Dataset class.

```python
class ComVEDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)
```

Turn encodings and labels into a Dataset object.

```python
train_dataset = ComVEDataset(train_encodings, train_labels)
# val_dataset = ComVEDataset(val_encodings, val_labels)
# test_dataset = ComVEDataset(test_encodings, test_labels)
```

## 3. Load model and optimizer

```python
from transformers import DistilBertForMultipleChoice, AdamW

model = DistilBertForMultipleChoice.from_pretrained('distilbert-base-uncased')
model.to(device)

optim = AdamW(model.parameters(), lr=5e-5)
```

## 4. Training

Prepare some utility functions.

```python
# Prediction function
def predict(outputs):
    pred = outputs["logits"].softmax(dim=1).argmax(dim=1, keepdim=True)
    return pred
```

Training procedure

```python
# Training
from torch.utils.data import DataLoader

# Set the number of epochs
epochs = 1

# Get training data by DataLoader
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)

# Start training
model.train()

train_loss = list()
for epoch in range(epochs):
    print('Epoch %s/%s' % (epoch + 1, epochs))
    time.sleep(0.3)
    
    correct = 0
    count = 0
    epoch_loss = list()
    
    pbar = tqdm(train_loader)
    for batch in pbar:
        optim.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs[0]
        loss.backward()
        optim.step()
        
        # make a prediction
        pred = predict(outputs)

        # count accuracy
        correct += pred.eq(labels.view_as(pred)).sum().item()
        count += len(labels)

        # show progress along with metrics
        pbar.set_postfix({
            'loss': '{:.3f}'.format(loss.item()),
            'acc': '{:.3f}'.format(correct * 1.0 / count)
        })
        
        # record the loss
        epoch_loss.append(loss.item())
        
    pbar.close()
    train_loss += epoch_loss
```

Visualise training loss.

```python
import matplotlib.pyplot as plt
plt.plot(train_loss, label="Training Loss")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Iterations vs Loss")  
plt.legend()
plt.show()
```

## 5. Test (todo)

Test procedure

```python
# Testing
from torch.utils.data import DataLoader

# Get test data by DataLoader
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

# Start testing
model.eval()

with torch.no_grad():
    
    correct = 0
    count = 0
    record = list()
    
    pbar = tqdm(test_loader)
    for batch in pbar:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs[0]
        
        # make a prediction
        pred = predict(outputs)

        # count accuracy
        correct += pred.eq(labels.view_as(pred)).sum().item()
        count += len(labels)

        # show progress along with metrics
        pbar.set_postfix({
            'loss': '{:.3f}'.format(loss.item()),
            'acc': '{:.3f}'.format(correct * 1.0 / count)
        })
    
        # record the results
        record.append((int(pred),int(labels)))
        

    pbar.close()
```

Check the result.

```python
# Show test result
for idx,item in enumerate(record):
    print("Statements %s:" % idx)
    print("Model Prediction:",item[0],"vs. Ground Truth:",item[1])
```
